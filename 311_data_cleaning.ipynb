```python
import pandas as pd
```

```python
# read 311 data from 2016-1 to 2016-4
df311_20160104 = pd.read_csv("311_20160104.csv")
df311_20160104 = df311_20160104[["Created Date", "Complaint Type"]]
df311_20160104.head()
```

```python
df311_20160104.info()
```

```python
# Timestamp Clean
```

```python
df311_20160104['date2'] = df311_20160104['Created Date'].astype('datetime64')
df311_20160104.head()
```

```python
df311_20160104_monthly = df311_20160104.set_index("date2")
df311_20160104_monthly = df311_20160104_monthly.resample('M').count().reset_index()

df311_20160104_monthly.head()
```

```python
df311_20160104_monthly = df311_20160104_monthly[["date2", "Complaint Type"]]
df311_20160104_monthly.columns = ["Year_Month", "Number of Incidence"]
df311_20160104_monthly.head()
```
